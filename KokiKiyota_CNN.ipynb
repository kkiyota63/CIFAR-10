```python
!pip install torchinfo
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torchvision.datasets import CIFAR10
from torch.utils.data import DataLoader
from torchinfo import summary
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# GPUを利用できるかチェックし、デバイスを選択
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using', device)

# 畳み込みニューラルネットワークを定義
size_image = 32  # 訓練に利用する画像の長辺のサイズ（画素数、8～32程度の範囲に設定）
num_epochs = 10  # 訓練を繰り返す回数
size_batch = 100  # 1回の更新に使用する訓練サンプル数

# 畳み込み層のパラメータ
size_kernel1 = 3
size_kernel2 = 3
size_kernel3 = 3
size_pad1 = 1
size_pad2 = 1
size_pad3 = 1
stride1 = 1
stride2 = 1
stride3 = 1
channels1 = 16
channels2 = 32
channels3 = 64


# 畳み込みニューラルネットワークを定義
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(3, channels1, kernel_size=size_kernel1, padding=size_pad1, stride=stride1),
            nn.BatchNorm2d(channels1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(p=0.2)
        )
        self.layer2 = nn.Sequential(
            nn.Conv2d(channels1, channels2, kernel_size=size_kernel2, padding=size_pad2, stride=stride2),
            nn.BatchNorm2d(channels2),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(p=0.3)
        )
        self.layer3 = nn.Sequential(
            nn.Conv2d(channels2, channels3, kernel_size=size_kernel3, padding=size_pad3, stride=stride3),
            nn.BatchNorm2d(channels3),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(p=0.4)
        )
        size = int((size_image + size_pad1 * 2 - size_kernel1) / stride1 + 1)
        size = int(size / 2)
        size = int((size + size_pad2 * 2 - size_kernel2) / stride2 + 1)
        size = int(size / 2)
        size = int((size + size_pad3 * 2 - size_kernel3) / stride3 + 1)
        size = int(size / 2)
        self.fc = nn.Sequential(
            nn.Linear(size * size * channels3, 128),
            nn.Dropout(p=0.5),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.layer3(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x


# CNNモデルを作る
model = CNN().to(device)

# transformを定義
transform = transforms.Compose([
    transforms.Resize(size_image),
    transforms.ToTensor()
])


# CIFAR-10をダウンロード
train_dataset = CIFAR10(root='mydata', train=True, transform=transform, download=True)
test_dataset = CIFAR10(root='mydata', train=False, transform=transform)

#loadersを作る
train_loader = DataLoader(dataset=train_dataset, batch_size=size_batch, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=size_batch, shuffle=False)

# モデルのトレーニングに使用する最適化モジュールを定義
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# モデルを訓練する関数を定義
def train(train_loader):
    model.train()
    running_loss = 0
    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        running_loss += loss.item()
        loss.backward()
        optimizer.step()
    train_loss = running_loss / len(train_loader)
    return train_loss

def valid(test_loader):
    model.eval()
    running_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct += predicted.eq(labels).sum().item()
            total += labels.size(0)

    val_loss = running_loss / len(test_loader)
    val_acc = correct / total
    return val_loss, val_acc

loss_list = []
val_loss_list = []
val_acc_list = []

# 訓練済みモデルを出力
torch.save(model.state_dict(), "trained_cnn_model.pth")

for epoch in range(num_epochs):
    loss = train(train_loader)
    val_loss, val_acc = valid(test_loader)
    print('epoch %d, 各エポックでのトレーニングの損失: %.4f, バリデーションの損失: %.4f, バリデーションの精度: %.4f' % (epoch, loss, val_loss, val_acc))
    loss_list.append(loss)
    val_loss_list.append(val_loss)
    val_acc_list.append(val_acc)

# 損失値を図にする
plt.plot(range(num_epochs), loss_list, 'p-', label='train_loss')
plt.plot(range(num_epochs), val_loss_list, 'b-*', label='val_loss')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.figure()

# エポックごとの予測精度を図にする
plt.plot(range(num_epochs), val_acc_list, 'g-', label='val_acc')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.figure()

plt.show()

num_samples = 100  #予測する画像の数
correct_count = 0  # カウント用
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# テストデータの最初のバッチを取得
sample_inputs, sample_labels = next(iter(test_loader))

for id in range(num_samples):
    sample_input = sample_inputs[id]
    plt.imshow(np.transpose(sample_input.numpy(), (1, 2, 0)), interpolation='nearest')
    plt.show()

    print("Correct label: {}".format(classes[sample_labels[id].item()]))
    sample_input = sample_input.reshape(1, 3, size_image, size_image).to(device)
    outputs = model(sample_input)
    predicted_label = classes[outputs.max(1, keepdim=True)[1].item()]
    print("Predicted label: {}".format(predicted_label))

    if classes[sample_labels[id].item()] == predicted_label:
        correct_count += 1

#ここからは可視化用

misclassified_counts = {class_name: 0 for class_name in classes}

for id in range(num_samples):
    sample_input = sample_inputs[id]
    correct_label = classes[sample_labels[id].item()]

    sample_input = sample_input.reshape(1, 3, size_image, size_image).to(device)
    outputs = model(sample_input)
    predicted_label = classes[outputs.max(1, keepdim=True)[1].item()]

    if correct_label != predicted_label:
        misclassified_counts[correct_label] += 1

# Plot the misclassification counts as a bar graph
plt.bar(misclassified_counts.keys(), misclassified_counts.values())
plt.xlabel('Classes')
plt.ylabel('Number of misclassifications')
plt.title('Misclassifications for each class')
plt.xticks(rotation=45)  # Add this line to rotate x labels
plt.show()

# Switch the model to evaluation mode
model.eval()

# Create a list to store the true and predicted labels
true_labels = []
pred_labels = []

# Get the first batch of test data
sample_inputs, sample_labels = next(iter(test_loader))

for id in range(num_samples):
    sample_input = sample_inputs[id]
    sample_input = sample_input.reshape(1, 3, size_image, size_image).to(device)
    outputs = model(sample_input)
    predicted_label = outputs.max(1, keepdim=True)[1].item()

    true_labels.append(classes[sample_labels[id].item()])
    pred_labels.append(classes[predicted_label])

# Create a confusion matrix
cm = confusion_matrix(true_labels, pred_labels, labels=classes)

# Plot the confusion matrix
plt.figure(figsize=(10, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

#正解数と精度を表示
print("Total correct predictions: ", correct_count)
print("Accuracy: ", correct_count / num_samples * 100, "%")
```
